```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\meser\OneDrive\Desktop\Machine Learning\loan-test.csv")
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['Dependents'] = df['Dependents'].fillna(0)
```

```python
df
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna('No')
```

```python
df
```

```python
df.describe()
```

```python
df
```

```python
Property_Area = {'Rural': 0, 'Semiurban': 1, 'Urban': 2}
df['Property_Area'] = df['Property_Area'].map(Property_Area)
```

```python
df
```

```python
df['ApplicantIncome'] = df['ApplicantIncome'].astype(float)
```

```python
df['CoapplicantIncome'] = df['CoapplicantIncome'].astype(float)
```

```python
Education = {'Graduate': 1, 'Not Graduate': 0}
df['Education'] = df['Education'].map(Education)
```

```python
df
```

```python
# I am changing NaN Credit_History to 1 as they have already taken loan and it seems that they had a good credit history before
df['Credit_History'] = df['Credit_History'].fillna(1)
```

```python
df['Loan_Status'] = df['Loan_Status'].fillna(1)
```

```python
df
```

```python
# Loan_Term - I replaced NaN values with the "Mean" 
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(343)
```

```python
df
```

```python
df['Total Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python
df
```

```python
cols = df.columns.tolist()
cols.insert(-5, cols.pop(cols.index('Total Income')))
df = df[cols]
```

```python
df
```

```python
df.columns = df.columns.str.replace(" ", "_").str.lower()  # Replace spaces with underscores and lowercase
df
```

```python
df['loanamount'] = df['loanamount'].fillna(136)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

x = df[['total_income', 'credit_history', 'loanamount', 'property_area']]  # Features
y = df['loan_status']  # Target

#total_income	loanamount	loan_amount_term	credit_history	property_area	loan_status
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=362)
```

```python
our_model = LogisticRegression(max_iter=1000)
our_model.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_pred = our_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')  # Use 'weighted' or 'macro' for multiclass
recall = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(our_model, x, y, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy: {scores.mean():.2f} (+/- {scores.std():.2f})")
```

```python
test_accuracy = our_model.score(x_test, y_test)
print(f"Test Accuracy: {test_accuracy:.2f}")
```
